```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt



sigmoid_data = pd.read_csv("sigmoid20171213-133649new.csv", index_col = 0)
#tanh_data = pd.read_csv("tanh20171211-215319.csv", index_col = 0)
```

```python
cols = ['error_A_m', 'error_A_m_1', 'error_V_m', 'error_V_m_1','max_dist_A_m', 'max_dist_A_m_1', 'max_dist_V_m',
       'max_dist_V_m_1', 'min_dist_A_m', 'min_dist_A_m_1', 'min_dist_V_m',
       'min_dist_V_m_1']
```

```python
def plotvalues(data1, data2, data3, m, title):
    ds = np.array(list(range(11)))*2 + m
    #index = (int(np.floor(i/2)), i % 2)
    plt.plot(ds, data1, "bx-", label="$m_1 = 1$")
    plt.plot(ds, data2, "rx-", label="$m_1 = 10$")
    plt.plot(ds, data3, "gx-", label="$m_1 = 20$")
    plt.xlabel(r"$d$")

    plt.ylabel(r"{}".format(title))
    plt.title(s="{} for m={}, m_1 = 1,10,20".format(title, m),  fontsize=16)
    plt.legend(loc='upper left')
    
```

```python
sigmoid_data.head()
```

```python
m = 40
#_, pltarr = plt.subplots(6,2)
for title in ["error_A_m_1"]:
    X1 = sigmoid_data[(sigmoid_data["m_1"] ==1) & (sigmoid_data["m"] ==m)][title]
    X2 = sigmoid_data[(sigmoid_data["m_1"] ==10) & (sigmoid_data["m"] ==m)][title]
    X3 = sigmoid_data[(sigmoid_data["m_1"] ==20) & (sigmoid_data["m"] ==m)][title]

    plotvalues(X1,X2,X3, m, title)
    
```

```python
m = 20
#_, pltarr = plt.subplots(6,2)
for title in ["error_V_m_1"]:
    X1 = sigmoid_data[(sigmoid_data["m_1"] ==1) & (sigmoid_data["m"] ==m)][title]
    X2 = sigmoid_data[(sigmoid_data["m_1"] ==10) & (sigmoid_data["m"] ==m)][title]
    X3 = sigmoid_data[(sigmoid_data["m_1"] ==20) & (sigmoid_data["m"] ==m)][title]

    plotvalues(X1,X2,X3, m, title)
```
